<p style="font-family: 'Times New Roman', serif; font-size: 35px;">
Solution of Cauchy problem for First-Order PDE
</p>

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Problem:</span> <span style="font-family: 'Times New Roman', serif; font-size: 28px">To obtain the solution of a Cauchy problem for the given first-order PDE</span>  
<span style="font-size:28px">
 $$  u_x - u_y = 1 $$
</span>  
<span style="font-family: 'Times New Roman', serif; font-size: 28px">with the Cauchy data $ u(x, 0) = x^2 $.</span>

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Solution:</span> 

```
pde = D[u[x, y], x] - D[u[x, y], y] == 1

initialCondition = u[x, 0] == x^2;

solution = DSolve[{pde, initialCondition}, u[x, y], {x, y}]

surface = Plot3D[
    Evaluate[u[x, y] /. solution], {x, -5, 5}, {y, -5, 5},
    AxesLabel -> {"x", "y", "u"},
    PlotLabel -> "Integral surface through initial curve"
];

initialCurve = ParametricPlot3D[
    {t, 0, t^2}, {t, -5, 5},
    PlotStyle -> {Thick, Red}
];

Show[surface, initialCurve]
```

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Result:</span> <span style="font-family: 'Times New Roman', serif; font-size: 28px">The solution of the given Cauchy problem for the given PDE is obtained in the output and visualized as an integral surface.</span>

<p style="font-family: 'Times New Roman', serif; font-size: 35px">Plotting the Characteristics for the First-Order PDE</p>

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Problem:</span> <span style="font-family: 'Times New Roman', serif; font-size: 28px">To find the characteristics of the equation</span>  
<span style="font-size:28px">
 $$ \left (u - y \right )\,u_x + y \,u_y = x + y $$
</span>  <span style="font-family: 'Times New Roman', serif; font-size: 28px">and plot them.</span>

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Solution:</span> 

<span style="font-family: 'Times New Roman', serif; font-size: 28px"> The auxiliary equations of the given PDE is</span> <br>
<span style="font-size:28px">
 $$  \dfrac{dx}{u - y} = \dfrac{dy}{y} = \dfrac{du}{x + y} \tag{1} $$
</span>
<span style="font-family: 'Times New Roman', serif; font-size: 28px">We have,</span>  
<span style="font-size:28px">
 $$ \Rightarrow \dfrac{dx + du}{(u - y) + (x + y)} = \dfrac{dy}{y} $$
 $$ \Rightarrow \dfrac{d(x + u)}{x + u} = \dfrac{dy}{y} $$
 $$ \Rightarrow \int \dfrac{d(x + u)}{x + u} = \int \dfrac{dy}{y} $$
 $$ \Rightarrow \log{(x + u)} = \log{(y)} + \log{(c_1)} $$
 $$ \Rightarrow \dfrac{x + u}{y} =  c_1 $$
 $$ \therefore\; u = y\,c_1 - x \tag{2} $$
</span>
<span style="font-family: 'Times New Roman', serif; font-size: 28px">Here, $ c_1 $ is a constant. <br>
Again, we have from Eq.(1),</span>
<span style="font-size:28px">
 $$ \Rightarrow \dfrac{dx + dy}{(u - y) + y} = \dfrac{du}{x+y} $$
 $$ \Rightarrow \dfrac{d(x + y)}{u} = \dfrac{du}{x+y} $$
 $$ \Rightarrow (x+y)\,d(x+y) = u\,du $$
 $$ \Rightarrow \int (x+y)\,d(x+y) = \int u\,du $$
 $$ \Rightarrow \dfrac{(x+y)^2}{2} = \dfrac{u^2}{2} + \dfrac{c_2}{2} $$
 $$ \Rightarrow (x+y)^2 - u^2 = c_2 $$
 $$ \therefore\; u = \sqrt{\left(x+y\right)^2 - c_2} \tag{3}$$
</span>  
<span style="font-family: 'Times New Roman', serif; font-size: 28px">Here, $ c_2 $ is a constant.

```
(* for c1 = 0, 5, 10 *)
f0 = Plot3D[-x, {x, -5, 5}, {y, -5, 5},
    AxesLabel -> {"x", "y", "u"},
    PlotLabel -> "Characteristics of Eq.(2) for c1 = 0, 5, 10"
];
f1 = Plot3D[5*y - x, {x, -5, 5}, {y, -5, 5}];
f2 = Plot3D[10*y - x, {x, -5, 5}, {y, -5, 5}];
    
Show[f0, f1, f2]

(* for c2= 0, 5, 10 *)
h0 = Plot3D[x + y, {x, -5, 5}, {y, -5, 5},
    AxesLabel -> {"x", "y", "u"},
    PlotLabel -> "Characteristics of Eq.(3) for c2 = 0, 5, 10"
];
h1 = Plot3D[Sqrt[(x + y)^2 + 5], {x, -5, 5}, {y, -5, 5}];
h2 = Plot3D[Sqrt[(x + y)^2 + 10], {x, -5, 5}, {y, -5, 5}];

Show[h0, h1, h2]
```

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Result:</span> <span style="font-family: 'Times New Roman', serif; font-size: 28px">The characteristics of the given PDE are plotted in the output.</span>

<p style="font-family: 'Times New Roman', serif; font-size: 35px">Plot the Integral Surfaces of a given First-Order PDE with Initial Data</p>

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Problem:</span> <span style="font-family: 'Times New Roman', serif; font-size: 28px">To find the solution of the given PDE</span>  
<span style="font-size:28px">
 $$  u_x + y\, u_y = 0 $$
</span> 
<span style="font-family: 'Times New Roman', serif; font-size: 28px">with the Cauchy data  $ u(0, y) = 4y $. </span>  <br> 
<span style="font-family: 'Times New Roman', serif; font-size: 28px">Plot the integral surface with in the range  $ (x, -5, 5) $ & $ (y, -5, 5) $.</span>

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Solution:</span> 

```
pde = D[u[x, y], x] + y*D[u[x, y], y] == 0

initialCondition = u[0, y] == 4*y;

solution = DSolve[{pde, initialCondition}, u[x, y], {x, y}]

Plot3D[
     Evaluate[u[x, y] /. solution], {x, -5, 5}, {y, -5, 5},
     AxesLabel -> {"x", "y", "u"},
     PlotLabel -> "Integral surface"
]
```

<span style="font-family: 'Times New Roman'; font-weight: bold; font-size: 28px">Result:</span> <span style="font-family: 'Times New Roman', serif; font-size: 28px">The solution of the given Cauchy problem for the given PDE is obtained in the output and visualized as an integral surface.</span>
